```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
le=LabelEncoder()
oe=OrdinalEncoder()
```

```python
train1=train.iloc[:,:-1].apply(le.fit_transform)
train1
```

```python
train1['Salary']=train['Salary']
train1.head()
```

```python
test1=test.iloc[:,:-1].apply(le.fit_transform)
test1
```

```python
test1['Salary']=train['Salary']
test1.head()
```

```python
# splitting train and test data
```

```python
xtrain=train1.iloc[:,:13]
ytrain=train1['Salary']
```

```python
xtest=test1.iloc[:,:13]
ytest=test1['Salary']
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

# Model building

```python
def norm(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
xtrain=norm(xtrain)
xtest=norm(xtest)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.svm import SVC
```

```python
# linear model
linear=SVC(kernel='linear')
linear.fit(xtrain,ytrain)
```

```python
ypred=linear.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```

```python
# poly model
poly=SVC(kernel='poly')
poly.fit(xtrain,ytrain)
```

```python
ypred1=poly.predict(xtest)
```

```python
accuracy_score(ytest,ypred1)
```

```python
# sigmoid model
sigm=SVC(kernel='sigmoid')
sigm.fit(xtrain,ytrain)
```

```python
ypred2=sigm.predict(xtest)
```

```python
accuracy_score(ytest,ypred2)
```

```python
# rbf model
rbf=SVC(kernel='rbf')
rbf.fit(xtrain,ytrain)
```

```python
ypred3=rbf.predict(xtest)
```

```python
accuracy_score(ytest,ypred3)
```
